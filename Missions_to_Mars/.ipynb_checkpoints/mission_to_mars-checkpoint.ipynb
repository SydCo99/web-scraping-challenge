```python
# Dependencies
from bs4 import BeautifulSoup
import requests
import os
from splinter import Browser
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
from pprint import pprint
```

Part 1: Scraping

```python
#Scrape the [Mars News Site](https://redplanetscience.com/) and collect the latest News Title and Paragraph Text. 

executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

1.1 NASA Mars News


```python
url = 'https://redplanetscience.com/'
browser.visit(url)
```

```python
html = browser.html
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
soup
```

```python
# Assign the text to variables that you can reference later.
news_title = soup.find_all('div', class_='content_title')[0].text

news_paragraph = soup.find_all('div', class_='article_teaser_body')[0].text

print(f"{news_title} : {news_paragraph}")
```

1.4 Mars Hemispheres

```python
### Mars Hemispheres

url = "https://marshemispheres.com/"
browser.visit(url)
html = browser.html
```

```python
hemisphere_soup = BeautifulSoup(html, 'html.parser')
```

```python
all_hemispheres = hemisphere_soup.find('div', class_='collapsible results')
queries = all_hemispheres.find_all('div', class_= 'item')
queries
```

```python
final_dict = []
    
for query in queries: 
    hemisphere = query.find('div', class_="description")
    hemisphere_title = hemisphere.h3.text

    img_url = query.find("a")
    hemisphere_dict = {}
    hemisphere_dict["title"] = hemisphere_title 
    hemisphere_dict["img_url"] = img_url 
    

final_dict.append(hemisphere_dict)
final_dict
```

```python
hemisphere_titles = all_hemispheres.find_all('h3')

titles = []
for title in range(len(hemisphere_titles)):
    titles.append(hemisphere_titles[title].text)
titles 
```
